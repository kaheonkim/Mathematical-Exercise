# Fourier sine series
$$\phi(x)=\sum_{n=1}^\infty A_n\sin(\frac{n\pi x}{l})$$

in the interval $[0,l]$.


By the property of integration 
$$\int_{0}^l\sin(\frac{n\pi x}{l})\sin(\frac{m\pi x}{l})dx = 0$$ if $n\neq m$ 

we can derive $A_m$ as
$$A_m = \frac{2}{l}\int_0^l\phi(x)\sin(\frac{m\pi x}{l})dx$$

```python
#Required Package
import numpy as np
import matplotlib.pyplot as plt
```

## Fourier sine series for $\phi(x)=1$ in the interval $[0,l]$
By the equation 

$$A_m = \frac{2}{l}\int_0^l \sin(\frac{m\pi x}{l})dx 
=\frac{2}{m\pi}(1-\cos(m\pi))
= \begin{cases}\frac{4}{m\pi} &&\text{if n is odd}\\
0 &&\text{if n is even}\end{cases}$$

```python
l = 10
def fourier_sine_1(x,num):
    func = 0
    for i in range(num):
        func += 4/(np.pi)/(2*i+1)*np.sin(x*(2*i+1)*np.pi/l)
        
    return func
```

```python
x = np.linspace(1,10,1000)
y = fourier_sine_1(x, 100)
plt.plot(x,y,'r')
plt.plot(x,[1]*len(x),'b')
```

## Fourier sine series for $\phi(x)=x$ in the interval $[0,l]$
By the equation 

$$A_m = \frac{2}{l}\int_0^l x\sin(\frac{m\pi x}{l})dx 
=-\frac{2l}{m\pi}\cos(m\pi)
= (-1)^{m+1}\frac{2l}{m\pi}$$

```python
l = 10
def fourier_sine_x(x,num):
    func = 0
    for i in range(1,num+1):
        func += (-1)**(i+1)*2*l/(i*np.pi)*np.sin(i*np.pi*x/l)
    return func
```

```python
x = np.linspace(1,10,1000)
plt.plot(x,fourier_sine_x(x, 1000),'r')
plt.plot(x,x,'b')
```

## Fourier sine series for $\phi(x)=x^2$ in the interval $[0,l]$
By the equation 

$$A_m = \frac{2}{l}\int_0^l x^2\sin(\frac{m\pi x}{l})dx 
=\frac{2}{l}(-\frac{l^3}{m\pi}\cos(m\pi)+\frac{2l^3}{(m\pi)^3}(\cos(m\pi)-1))
= \begin{cases}\frac{2}{l}(\frac{l^3}{m\pi}+\frac{-4l^3}{(m\pi)^3}) && \text{m : odd}\\
\frac{2}{l}(-\frac{l^3}{m\pi})&& \text{m : even}\end{cases}$$

```python
l=1
def fourier_sine_x_sq(x,num):
    func = 0
    for i in range(1,num+1):
        if i % 2 ==0:
            func += 2/l*(-l**3/i/np.pi)*np.sin(i*np.pi*x/l)
        else:
            func += 2/l*(l**3/i/np.pi-4*l**3/(i*np.pi)**3)*np.sin(i*np.pi*x/l)
    return func
```

```python
x = np.linspace(0,1,1000)
y = fourier_sine_x_sq(x, 1000)
plt.plot(x,y,'r')
plt.plot(x,x**2,'b')
```

# Fourier cosine series
$$\phi(x)=\frac{1}{2}A_0+\sum_{n=1}^\infty A_n\cos(\frac{n\pi x}{l})$$

in the interval $[0,l]$.


By the similar property of integration, we can derive $A_m$ as
$$A_m = \frac{2}{l}\int_0^l\phi(x)\cos(\frac{m\pi x}{l})dx$$

## Fourier cosine series for $\phi(x)=x$ in the interval $[0,l]$
By the equation 
$$A_0=\frac{2}{l}\int_0^lxdx =l$$
$$A_m = \frac{2}{l}\int_0^l x \cos(\frac{m\pi x}{l})dx 
=\frac{2l}{m^2\pi^2}((-1)^m-1)
= \begin{cases}-\frac{4l}{m^2\pi^2} &&\text{if m is odd}\\
0 &&\text{if m is even}\end{cases}$$

```python
l = 10
def fourier_cosine_x(x,num):
    func = l/2
    for i in range(num):
        if i %2 ==1:
            func += (-1)*4*l/(i*np.pi)**2*np.cos((i)*np.pi*x/l)
    return func
```

```python
y = fourier_cosine_x(x,1000)
plt.plot(x,y,'r')
plt.plot(x,x,'b')
```

## Fourier cosine series for $\phi(x)=x^2$ in the interval $[0,l]$
By the equation 

$$A_m = \frac{2}{l}\int_0^l x^2\cos(\frac{m\pi x}{l})dx 
=\frac{4l^2}{(m\pi)^2}(\cos(m\pi))
= \begin{cases}\frac{-4l^2}{(m\pi^2)} && \text{m : odd}\\
\frac{4l^2}{(m\pi^2)}&& \text{m : even}\end{cases}$$

```python
l=1
def fourier_cosine_x_sq(x,num):
    func = l**2/3
    for i in range(1,num+1):
        if i % 2 == 0:
            func += 4*l**2/i**2/np.pi**2*np.cos(i*np.pi*x/l)
        if i % 2 == 1:
            func += -4*l**2/i**2/np.pi**2*np.cos(i*np.pi*x/l)
    return func
```

```python
x = np.linspace(0,1,100)
y = fourier_cosine_x_sq(x, 100)
plt.plot(x,y,'r')
plt.plot(x,x**2,'b')
```

# Fourier Full series
$$\phi(x)=\frac{1}{2}A_0+\sum_{n=1}^\infty (A_n\cos(\frac{n\pi x}{l})+B_n\sin(\frac{n\pi x}{l}))$$

in the interval $[-l,l]$.


By combining coefficients for Fourier cosine series and Fourier sine series,
$$A_n = \frac{1}{l}\int_{-l}^l\phi(x)\cos(\frac{n\pi x}{l})dx$$
$$B_n = \frac{1}{l}\int_{-l}^l\phi(x)\sin(\frac{n\pi x}{l})dx$$

## Fourier full series for $\phi(x)=x$ in the interval $[-l,l]$
By the equation 
$$A_0=\frac{1}{l}\int_{-l}^lxdx =0$$
$$A_m = \frac{1}{l}\int_{-l}^l x \cos(\frac{m\pi x}{l})dx 
=\frac{l}{m^2\pi^2}(\cos(m\pi)-\cos(-m\pi))
= 0$$
$$B_m = \frac{1}{l}\int_{-l}^l x \sin(\frac{m\pi x}{l})dx 
=\frac{-2l}{m\pi}\cos(m\pi)=(-1)^{m+1}\frac{2l}{m\pi}$$

```python
l = 10
def fourier_full_x(x,num):
    func = 0
    for i in range(1,num+1):
        if i % 2 ==0:
            func += -2*l/(i*np.pi)*np.sin(i*np.pi*x/l)
        else:
            func += 2*l/(i*np.pi)*np.sin(i*np.pi*x/l)
    return func
```

```python
y = fourier_full_x(x,1000)
plt.plot(x,y,'r')
plt.plot(x,x,'b')
```
